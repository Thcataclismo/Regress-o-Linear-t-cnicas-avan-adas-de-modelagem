# **Regressao linear com SCIKIT LEARN**


### 1-Estimando o modelo com os dados de treino

```python
#Primeiramente, iremos importar as bibliotecas a serem utilizadas
from sklearn.linear_model import LinearRegression
from sklearn import metrics 
```

```python
#faremos a instanciação da classe LinearRegeression().
modelo =  LinearRegression() 
```

```python
#utilizaremos o método fit() para realizar o ajuste e estimar nosso modelo linear utilizando os dados de treino.
modelo.fit(X_train, y_train)
```

```python
#O que faremos agora é obter R² utilizando o sckitlearn, e faremos um arredondamento dos valores ao utilizar o round() :
print('R² = {}'.format(modelo.score(X_train, y_train).round(3)))
```

```python
#Criaremos uma variável y_previsto
y_previsto = modelo.predict(X_test)
```

```python
#obteremos o coeficiente de determinações para as previsões do nosso modelo
print('R² = %s' % metrics.r2_score(y_test, y_previsto).round(3))
```

### 2-Obtendo previsões pontuais

```python
#Continuaremos a usar os dados de teste, que serão armazenados na variável entrada, e coletaremos o primeiro registro do dataset.
entrada = X_test[0:1]
```

```python
#Para gerar a previsão, chamaremos o modelo.predict, passaremos a entrada e acrescentaremos [0] para indicar que queremos o primeiro elemento do array
modelo.predict(entrada)[0]
```

```python
# precisamos inverter essa transformação para obter o valor em reais
np.exp(modelo.predict(entrada)[0]) 
```

```python
#Em seguida, geraremos um simulador simples em que podemos utilizar quaisquer valores que sejam do nosso interesse. 
#Lembrando que os valores passados pelo usuário não estarão em log, então precisaremos realizar essa transformação posteriormente.
Area = 250
Dist_Praia = 1
entrada = [[np.log(Area), np.log(Dist_Praia + 1)]]

print('R$ {0: 2f}'.format(np.exp(modelo.predict(entrada)[0])))
```

### 3-Interpretação dos coeficientes estimados

```python
#Para obter o intercepto, escreveremos:modelo.intercept_
modelo.intercept_
```

```python
#Faremos a transformação do valor obtido em reais, ao escrever
np.exp(modelo.intercept_)
```

```python
#Agora começaremos a obter os coeficientes de regressão, também conhecidos como coeficientes parciais angulares. Esse tipo de modelo é muito utilizado em estudos econômicos, pois eles medem o que chamamos de elasticidade da variável dependente em relação às variáveis explicativas
modelo.coef_
```

```python
#Criaremos uma pequena coluna para confirmar a ordem das variáveis explicativas no dataframe:
X.columns
```

```python
# criaremos um dataframe para guardar as informações dos coeficientes e exibir uma tabela no Jupyter.
index = ['Intercepto','log Área', 'log Distância até a Praia']
```

```python
#Exibiremos um dataframe para armazenar os coeficientes do modelo:pd.DataFrame(data=np.append(modelo.intercept_, modelo.coef_), index=index, columns=['Parâmetros'])
pd.DataFrame(data=np.append(modelo.intercept_, modelo.coef_), index=index, columns=['Parâmetros'])
```
